```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df
```

```python
df.isna().sum()
```

# Handling Missing Values

```python
df['Item_Weight']=df['Item_Weight'].fillna(df['Item_Weight'].mean())
df['Item_Weight'].isna().any()
```

```python
df['Outlet_Size'].mode()
```

```python
df['Outlet_Size']=df['Outlet_Size'].fillna('Medium')
```

```python
df.isna().any()
```

# Dealing with Categorical data

```python
df['Item_Fat_Content'].unique()
```

```python
a={'Low Fat':0,'Regular':1,'low fat':0,'LF':0,'reg':1}
```

```python
df['Item_Fat_Content']=df['Item_Fat_Content'].map(a)
df['Item_Fat_Content']
```

```python
df['Item_Type'].unique()
```

```python
dummy=pd.get_dummies(df['Item_Type'],drop_first=True)
dummy
```

```python
df=pd.concat([df,dummy],axis=1)
df
```

```python
df=df.drop(columns=['Item_Type'])
df
```

```python
df['Outlet_Size']=df['Outlet_Size'].map({'Small':0,'Medium':1,'High':2})
df['Outlet_Size']
```

```python
dummy1=pd.get_dummies(df[['Outlet_Location_Type','Outlet_Type']],drop_first=True)
dummy1
```

```python
df=pd.concat([df,dummy1],axis=1)
df
```

```python
df.dtypes
```

# Standardising the columns

```python
minIW=df['Item_Weight'].min()
rangeIW=df['Item_Weight'].max()-df['Item_Weight'].min()
df['Item_Weight']=(df['Item_Weight']-minIW)/rangeIW
df['Item_Weight']
```

```python
minEY=df['Outlet_Establishment_Year'].min()
rangeEY=df['Outlet_Establishment_Year'].max()-df['Outlet_Establishment_Year'].min()
df['Outlet_Establishment_Year']=(df['Outlet_Establishment_Year']-minEY)/rangeEY
df['Outlet_Establishment_Year']
```

```python
X=df.drop(columns=['Item_Identifier','Outlet_Identifier','Item_Outlet_Sales'])
y=df['Item_Outlet_Sales']
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```
