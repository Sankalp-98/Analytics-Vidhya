```python
import numpy as np
import pandas as pd
```

```python
df_train=pd.read_csv('Train.csv')
df_test=pd.read_csv('Test.csv')
```

```python
df_train
```

```python
df_train.drop(columns=['ID','Open','High','Low'],inplace=True)
df_train
```

```python
df_test
```

```python
df_test.drop(columns=['ID'],inplace=True)
```

```python
df_train['Date']=pd.to_datetime(df_train['Date'])
df_train['Date'][0]
```

```python
df_test['Date']=pd.to_datetime(df_test['Date'])
```

```python
df_train['Date'][0].day
```

```python
Year=[]
Month=[]
Day=[]
for i in range(len(df_train)):
    Year.append(df_train['Date'][i].year)
    Month.append(df_train['Date'][i].month)
    Day.append(df_train['Date'][i].day)
```

```python
df_train['Year']=Year
df_train['Month']=Month
df_train['Day']=Day

df_train
```

```python
df_train.drop(columns='Date',inplace=True)
```

```python
Year=[]
Month=[]
Day=[]
for i in range(len(df_test)):
    Year.append(df_test['Date'][i].year)
    Month.append(df_test['Date'][i].month)
    Day.append(df_test['Date'][i].day)
```

```python
df_test['Year']=Year
df_test['Month']=Month
df_test['Day']=Day

df_test
```

```python
df_test.drop(columns='Date',inplace=True)
```

```python
df_train
```

```python
df_test
```

```python
target=df_train['Close']
df_train.drop(columns='Close',inplace=True)
df_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
train_scaled=pd.DataFrame(sc.fit_transform(df_train))
train_scaled.columns=df_train.columns
train_scaled.index=df_train.index
```

```python
train_scaled
```

```python
X_test=pd.DataFrame(sc.fit_transform(df_test))
X_test.columns=df_test.columns
X_test.index=df_test.index
```

```python
X_test
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_valid,y_train,y_valid=train_test_split(train_scaled,target,test_size=0.2,random_state=24)
```

```python
X_train.shape
```

```python
X_valid.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
predict=lr.predict(X_valid)
predict
```

```python
y_valid
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from math import sqrt
```

```python
sqrt(mean_squared_error(predict,y_valid))
```

```python
submissions=pd.read_csv('SampleSubmission.csv')
submissions['Close']=lr.predict(X_test)
```

```python
submissions.to_csv('LinearReg.csv',index=False)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(X_train,y_train)
```

```python
predict=dt.predict(X_valid)
```

```python
sqrt(mean_squared_error(y_valid,predict))
```

```python
dt.fit(train_scaled,target)
```

```python
submissions['Close']=dt.predict(X_test)
submissions.to_csv('DecisionTree1.csv',index=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor(n_estimators=1000)
rf.fit(X_train,y_train)
```

```python
predict=rf.predict(X_valid)
sqrt(mean_squared_error(y_valid,predict))
```

```python
submissions['Close']=rf.predict(X_test)
submissions.to_csv('RandomForest.csv',index=False)
```
