```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

# Importing Train dataset

```python
df=pd.read_csv('train.csv')
df
```

```python
df.isna().sum()
```

# Dealing with missing values

```python
df1=df.fillna(df['Number_Weeks_Used'].median())
df1
```

```python
df1.isna().any()
```

```python
X=df1.drop(columns=['ID','Crop_Damage'])
X
```

```python
y=df1['Crop_Damage']
y
```

```python
X.describe()
```

# Scaling X

```python
s=MinMaxScaler()
```

```python
new_X=pd.DataFrame(s.fit_transform(X))
```

```python
new_X.columns=X.columns
new_X
```

```python
new_X.describe()
```

```python
y.describe()
```

```python
X_train,X_test,y_train,y_test=train_test_split(new_X,y,test_size=0.2)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
prediction=pd.DataFrame({'pred':prediction})
prediction['pred'].value_counts()
```

```python
model.score(X_test,y_test)
```

```python
final_model=GaussianNB()
```

```python
final_model.fit(new_X,y)
```
