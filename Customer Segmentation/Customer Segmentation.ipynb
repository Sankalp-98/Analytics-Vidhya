```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Train.csv')
df
```

```python
df.isna().sum()
```

```python
df1=df.dropna()
df1
```

```python
df1['Profession'].unique()
```

```python
df1['Segmentation'].unique()
```

```python
X=df1.drop(columns=['ID','Segmentation'])
X
```

```python
new_X=pd.get_dummies(X)
new_X
```

```python
y=df1['Segmentation'].map({'A':1,'B':2,'C':3,'D':4})
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s1=StandardScaler()
```

```python
new=s1.fit_transform(new_X)
```

```python
new=pd.DataFrame(new)
new
```

```python
new.columns=new_X.columns
```

```python
new.index=new_X.index
```

```python
new
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(new,y,test_size=0.2)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(multi_class='ovr')
```

```python
lr.fit(X_train,y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,y_test)
```

# NaiveBayes

```python
X_train,X_test,y_train,y_test=train_test_split(new_X,y,test_size=0.2,random_state=21)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
nb=MultinomialNB()
```

```python
nb.fit(X_train,y_train)
```

```python
nb.predict(X_test)
```

```python
nb.score(X_test,y_test)
```

```python
from sklearn.naive_bayes  import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
gnb.score(X_test,y_test)
```
